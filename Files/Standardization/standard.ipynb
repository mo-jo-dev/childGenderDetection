```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_train = pd.read_csv('../X_Train.csv')
df_test = pd.read_csv('../X_Test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
scaler = StandardScaler()
scaler.fit(df_train)
scaled_data_train = scaler.transform(df_train)
```

```python
scaler = StandardScaler()
scaler.fit(df_test)
scaled_data_test = scaler.transform(df_test)
```

```python
scaled_data_train_df = pd.DataFrame(scaled_data_train)
scaled_data_train_df.head()
```

```python
scaled_data_test_df = pd.DataFrame(scaled_data_test)
scaled_data_test_df.head()
```

```python
scaled_data_train_df.to_csv('X_Train.csv', index = 0)
scaled_data_test_df.to_csv('X_Test.csv', index = 0)
```
