```python
import librosa
from tqdm import tqdm
import numpy as np
import pandas as pd
```

```python
def feature_extraction(file_path):
    y, sample_rate = librosa.load(file_path, res_type = "soxr_hq")
    mfcc = np.mean(librosa.feature.mfcc(y = y, sr = sample_rate, n_mfcc = 39).T, axis = 0)

    return mfcc
```

```python
features = {}
```

```python
with open("path_train.txt", "r") as file:
    audio_paths = file.readlines()

audio_paths = [ path.strip() for path in audio_paths ]
```

```python
for path in tqdm(audio_paths):
    features[path] = feature_extraction(path)
```

```python
df = pd.DataFrame.from_dict(features.values())
df
```

```python
df.to_csv('X_Train.csv', index = False)
```
