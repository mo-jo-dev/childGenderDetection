```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.98)
```

```python
df_train = pd.read_csv('../Standardization/X_Train.csv')
df_test = pd.read_csv('../Standardization/X_Test.csv')
```

```python
df_train.head()
```

```python
df_test.head()
```

```python

df_train_pca = pca.fit_transform(df_train)
df_train = pd.DataFrame(df_train_pca)
df_train.head()
```

```python
df_train.to_csv('X_Feature_Train.csv', index=0)
```

```python
df_test_pca = pca.transform(df_test)
df_test = pd.DataFrame(df_test_pca)
df_test.head()
```

```python
df_test.to_csv('X_Feature_Test.csv', index=0)
```
