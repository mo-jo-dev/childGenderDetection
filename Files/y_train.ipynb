```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("path_train.txt", header = None)
df.columns = ['Name']
df
```

```python
x = df.Name.str[16]
y = df.Name.str[17:19]
df['Gender'] = x
df['Age'] = y
```

```python
gender_labels = df['Gender']
encoder = LabelEncoder()
z = encoder.fit_transform(gender_labels)
df['Gender_bin'] = z
df
```

```python
df.to_csv("y_train.csv", index = False)
```
